## Libraries

```python
import pandas as pd
import pandas_profiling
```

## Reading the *csv* file that has the consolidated data. 

```python
file = pd.read_csv('india.csv')
```

## Creating Dataframes for Hindi as the first, second, third and fourth Language option for the state

```python
Hindi_L1 = file[file['Language 1'].str.contains('Hindi')]
Hindi_L2 = file[file['Language 2'].str.contains('Hindi')]
Hindi_L3 = file[file['Language 3'].str.contains('Hindi')]
Hindi_L4 = file[file['Language 4'].str.contains('Hindi')]
```

## Calculating Population fraction based on Language preference for Hindi.

```python
Hindi_L1['Population'].sum()/file['Population'].sum(), Hindi_L2['Population'].sum()/file['Population'].sum(), Hindi_L3['Population'].sum()/file['Population'].sum(), Hindi_L4['Population'].sum()/file['Population'].sum()
```

## Profiling data

```python
pandas_profiling.ProfileReport(file)
```
